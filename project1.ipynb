```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
df =pd.read_csv(r"C:\Users\dell\Downloads\personal_expense_dataset.csv")
print(df.head())
```

```python
df.drop(columns=['Food_Ratio','Groceries_Ratio','Transport_Ratio','Entertainment_Ratio','Shopping_Ratio','Rent_Ratio','Bills_Ratio','Education_Ratio','Healthcare_Ratio']) 
```

```python
df.info()
df.describe()
```

```python
festival_expense = df.groupby('Festivals')[['Year','Month']] .sum()

festival_expense.plot(kind='barh', figsize=(10,5))

plt.title("Festival Wise Spending")
plt.xlabel("Amount $")
plt.ylabel("Festival")
plt.show()
```

```python
category_expense = df.groupby('Year')[['Food' ,'Groceries' ,'Transport','Entertainment','Shopping','Rent','Bills','Healthcare','Education']].sum()

category_expense.plot(kind='barh')

plt.legend(
    loc='upper left',
    bbox_to_anchor=(1,1))
plt.title("Category Wise Expense")
plt.ylabel("Year")
plt.xlabel("Amount $")

plt.show()
```

```python
category_expense = df.groupby('Year')[['Food' ,'Groceries' ,'Transport','Entertainment','Shopping','Rent','Bills','Healthcare','Education']].sum()

category_expense.plot(marker = '*',linestyle ="--")

plt.legend(
    loc='upper left',
    bbox_to_anchor=(1,1))
plt.title("Category Wise Expense, graph 2")
plt.ylabel("Amount $")
plt.xlabel("Year")
plt.show()
```

```python
food = df["Food"].sum()
gro = df["Groceries"].sum()
tran = df["Transport"].sum()
ent = df["Entertainment"].sum()
shop = df["Shopping"].sum()
rent = df["Rent"].sum()
Bill = df["Bills"].sum()
hc= df["Healthcare"].sum()
edu = df["Education"].sum()
expenses = pd.Series([food,gro,tran,ent,shop,rent,Bill,hc,edu],
                   index = ['Food','Groceries','Transport','Entertainment','Shopping','Rent','Bills','Healthcare','Education'])
expenses.plot(    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    figsize=(8,8)
)
plt.title("Category-wise Expense Distribution")

plt.ylabel('')

plt.show()
```

```python

monthly_expense = df.groupby('Month')[['Income','Total_Expenses']].sum()
k= (monthly_expense.sort_values(by = 'Total_Expenses',ascending=False))
monthly_expense.plot(kind='line', marker ='o')
plt.title("Monthly Spending Trend")
plt.ylabel("Amount $")
plt.legend(
    loc='upper left',
    bbox_to_anchor=(1,1))
plt.show()
```

```python
month_avg = df.groupby('Month')['Total_Expenses'].sum().mean()

print("Average Monthly Spending:", "$", month_avg)
```

```python
year_avg = df.groupby('Year')['Total_Expenses'].sum().mean()
print("Average Yearly Spending: $",year_avg)
```
